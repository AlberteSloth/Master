```julia
using Plots
L_0 = 2*10^-2
N = 2
# psi = (k_cat*C_AMPK)/((k_cat*C_AMPK)+v_P)
reg_i1(rho) = 1/(1+L_0*(rho/(1-rho))^N)

plot(reg_i1;
    legend = false,
    linewidth = 2,
    color = :red,
    xlims = (0,1),
    aspect_ratio = 1,
    fmt = :png,
    xlabel = "Ψ",
    ylabel = "v",
    title = "Inhibition 1"
)
```

```julia
function reg_a1(rho)
    1-(1/(1+L_0*(rho/(1-rho))^N))
end

plot(reg_a1;
    legend = false,
    linewidth = 2,
    color = :red,
    xlims = (0,1),
    aspect_ratio = 1,
    xlabel = "Ψ",
    ylabel = "v",
    title = "Activation 1"
)
```

```julia
L_0 = 50
reg_a2(rho) = 1/(1+L_0*((1-rho)/(rho))^N)

plot(reg_a2;
    legend = false,
    linewidth = 2,
    color = :red,
    xlims = (0,1),
    aspect_ratio = 1,
    xlabel = "Ψ",
    ylabel = "v",
    title = "Activation 2"
)
```

```julia
reg_i2(rho) = 1-(1/(1+L_0*((1-rho)/(rho))^N))

plot(reg_i2;
    legend = false,
    linewidth = 2,
    color = :red,
    xlims = (0,1),
    aspect_ratio = 1,
    xlabel = "Ψ",
    ylabel = "v",
    title = "Inhibition 2"
)
```

```julia
reg2(rho) = reg_i1(rho)+reg_a2(rho)
plot(reg2;
    legend = false,
    linewidth = 2,
    color = :red,
    xlims = (0,1)
)
```

```julia
L_0 = 2*10^-2 
N = 2
k_cat = 0.01 #s^-1
v_P = 2*10^-4 # mM/s
psi(C_AMPK) = (k_cat*C_AMPK)/((k_cat*C_AMPK)+v_P)
plot(psi;
    legend = false,
    linewidth = 2,
    color = :red,
    xlims = (0,1),
    aspect_ratio = 1
)
```

```julia

reg_AMPK(C_AMPK) = 1-(1/(1+L_0*(psi(C_AMPK)/(1-psi(C_AMPK)))^N))

plot(reg_AMPK;
    legend = false,
    linewidth = 2,
    color = :red,
    xlims = (0,1),
    aspect_ratio = 1
)
```
